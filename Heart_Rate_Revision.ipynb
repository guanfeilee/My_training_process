```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks
t = np.linspace(0,1)
signal = 0.5 * np.sin(2 * np.pi * 5 * t) + \
         (np.sin(2 * np.pi * 20 * t) > 0.9).astype(float) + \
         np.random.normal(0, 0.05, len(t))
peaks,_=find_peaks(signal,height=0.7,distance=10)
fig,ax=plt.subplots(figsize=(10,5),dpi=100)
ax.plot(t,signal,color='gray',alpha=0.6,label='RAW_ECG_Signal')
ax.plot(t[peaks],signal[peaks],"x",color='red',markersize=10,label='Detected_Heartbeats')
ax.set_title(r'Real_Time_Heart_Detection',fontsize=10)
ax.set_xlabel('Time(s)')
ax.set_ylabel('Amplitude(mV)')
ax.legend()
ax.grid(axis='y',linestyle='--',alpha=0.7)
plt.show
print(f"Dected_Heart_Beat:{len(peaks)}")
plt.savefig('Heart_Rate_Revision.png')
```
