```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
plt.figure(figsize=(6,4))
plt.plot(x,y,color="navy",linewidth=2,marker='s',label='Growth_Rate')
plt.title(r'Biomedical_Data_Anaiysis:$y=x^2$')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude(mV)')
plt.legend()
plt.grid(True,linestyle='--')
plt.show
plt.savefig('my_first_graph_made_by_matplotlib.pdf')
```
