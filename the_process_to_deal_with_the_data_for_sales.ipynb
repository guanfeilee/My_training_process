```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"D:\李冠霏\个人文件\Python\origin.csv")
df=df.fillna('none')
type(df)
df['quantity']=df['quantity'].abs()
df.loc[df['quantity']==0,"quantity"]=1
df.loc[df["product_id"]=="INVALID"]
df.drop_duplicates
df=df[df['product_id']!="INVALID"]
df["unit_price"]=df["unit_price"].astype(str).str.replace('$','',regex=False)
df["unit_price"]=pd.to_numeric(df["unit_price"],errors="coerce")
df["order_date"]=pd.to_datetime(df["order_date"],errors='coerce')
the_highest_price=df["unit_price"].quantile(0.99)
df=df[df["unit_price"]<=the_highest_price]
text_columns=['customer_name','city','payment_method','product_category']
for col in text_columns:
    df[col]=df[col].astype(str).str.strip()
df["city"]=df["city"].str.title()
df["payment_method"]=df["payment_method"].str.title()
df["total_amount"]=df["unit_price"]*df["quantity"]
df["order_year_month"]=df["order_date"].dt.to_period("M")
df["order_day_of_week"]=df["order_date"].dt.day_name()
df.to_excel(r'D:\李冠霏\个人文件\Python/finished.xlsx')
```
